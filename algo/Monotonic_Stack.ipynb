#Что такое монотонный стек?

Монотонный стек - это распространенная структура данных в информатике, которая поддерживает свои элементы в определенном порядке. В отличие от традиционных стеков, монотонные стеки гарантируют, что элементы внутри стека расположены в **возрастающем** или **убывающем** порядке в зависимости от времени их поступления. Чтобы получить монотонные стеки, мы должны применить операции push и pop в зависимости от того, хотим ли мы монотонно увеличивающийся стек или монотонно уменьшающийся стек.

## Монотонно увеличивающийся стек
Монотонно увеличивающийся стек - это стек, в котором элементы размещаются в порядке возрастания.. Каждый новый элемент, добавляемый в стек, больше или равен элементу, расположенному под ним. Если новый элемент меньше, мы удаляем элементы из верхней части стека, пока не найдем тот, который меньше или равен новому элементу, или пока стек не опустеет. Это гарантирует, что стек всегда остается в порядке возрастания.
```
Пример: 1, 3, 10, 15, 17
```

* Инициализируйте пустой стек.
* Выполняйте итерации по элементам и для каждого элемента:
  * пока стек не пуст, И вершина стека больше текущего элемента
    * извлеките элемент из стека
  * Помещаем текущий элемент в стек.
* В конце итерации стек будет содержать элементы в монотонно возрастающем порядке.

```
Arr[] = {1, 7, 9, 5}
```

````python
def monotonicIncreasing(nums):
    stack = []
    result = []

    # Traverse the array
    for num in nums:
        # While stack is not empty AND top of stack is more than the current element
        while stack and stack[-1] > num:
            # Pop the top element from the stack
            stack.pop()
        # Push the current element into the stack
        stack.append(num)

    # Construct the result array from the stack
    while stack:
        result.insert(0, stack.pop())

    return result

# Example usage:
nums = [3, 1, 4, 1, 5, 9, 2, 6]
result = monotonicIncreasing(nums)
print("Monotonic increasing stack:", result)
````

## Монотонно уменьшающийся стек:
Монотонно убывающий стек - это стек, в котором элементы размещаются в порядке убывания от низа к вершине. Каждый новый элемент, добавляемый в стек, должен быть меньше или равен элементу, расположенному под ним. Если новый элемент больше вершины стека, то мы удаляем элементы из вершины стека до тех пор, пока не найдем тот, который больше или равен новому элементу, или пока стек не опустеет. Это гарантирует, что стек всегда остается в порядке убывания.
```
Пример: 17, 14, 10, 5, 1
```

* Начните с пустого стека.
* Выполняйте итерации по элементам входного массива.
  * Пока стек не пуст, А вершина стека меньше текущего элемента:
    * извлеките элемент из стека
  * Помещаем текущий элемент в стек.
* После перебора всех элементов стек будет содержать элементы в монотонном порядке убывания.

```
Arr[] = {7, 5, 9, 4}
```

````python
def monotonicDecreasing(nums):
    stack = []
    result = []

    # Traverse the array
    for num in nums:
        # While stack is not empty AND top of stack is less than the current element
        while stack and stack[-1] < num:
            # Pop the top element from the stack
            stack.pop()

        # Construct the result array
        if stack:
            result.append(stack[-1])
        else:
            result.append(-1)

        # Push the current element into the stack
        stack.append(num)

    return result

# Example usage:
nums = [3, 1, 4, 1, 5, 9, 2, 6]
result = monotonicDecreasing(nums)
print("Monotonic decreasing stack:", result)
````

## Next Greater Element (NGE)

Следующий больший элемент для элемента x — это первый больший элемент справа от x в массиве. Элементы, для которых не существует большего элемента, считают следующий больший элемент равным -1.

```
Входные данные : arr[] = [4, 5, 2, 25]
Выходные данные : 4 –> 5
               5 –> 25
               2 –> 25
              25 –> -1
Объяснение: кроме 25, каждый элемент имеет элемент, больший, чем они, представленные справа. сторона

Ввод : arr[] = [ 13 , 7, 6 , 12 ]
Выход : 13 -> -1
                7 -> 12
                6 -> 12
               12 -> -1
Объяснение: 13 и 12 не содержат ни одного элемента, большего, чем они есть. на правой стороне
```

Идея состоит в том, чтобы использовать два цикла. Внешний цикл выбирает все элементы один за другим. Внутренний цикл ищет первый больший элемент для элемента, выбранного внешним циклом. Если найден больший элемент, то этот элемент печатается следующим, в противном случае печатается -1.


Выполните шаги, указанные ниже, чтобы реализовать идею:

* Перейдите по массиву от индекса 0 до конца .
  * Для каждого элемента начинается еще один цикл от индекса i+1 до конца.
    * Если во втором цикле найден больший элемент, выведите его и прервите цикл, иначе выведите -1.
   

````python
# Function to print element and NGE pair for all elements of list
def printNGE(arr):

    for i in range(0, len(arr), 1):

        next = -1
        for j in range(i+1, len(arr), 1):
            if arr[i] < arr[j]:
                next = arr[j]
                break

        print(str(arr[i]) + " -- " + str(next))


# Driver program to test above function
arr = [11, 13, 21, 3]
printNGE(arr)
````

## Найдите следующий больший элемент, используя стек

Идея состоит в том, чтобы хранить элементы, для которых нам нужно найти следующий больший элемент, в стеке, и при обходе массива, если мы находим больший элемент, мы соединяем его с элементами из стека до верхнего элемента стека. меньше текущего элемента.

* Нажмите первый элемент в стек.
* Выберите остальные элементы один за другим и выполните следующие шаги в цикле.
  * Пометить текущий элемент как следующий.
  * Если стек не пуст, сравните самый верхний элемент стека со следующим .
  * Если next больше верхнего элемента, извлеките элемент из стека. next — следующий больший элемент для вытолкнутого элемента.
  * Продолжайте извлекать из стека, пока вынутый элемент меньше следующего. next становится следующим большим элементом для всех таких выдвинутых элементов.
* Наконец, нажмите следующий в стеке.
* После завершения цикла на шаге 2 извлеките все элементы из стека и выведите -1 в качестве следующего элемента для них.

````python
# Python program to print next greater element using stack

# Stack Functions to be used by printNGE()


def createStack():
    stack = []
    return stack


def isEmpty(stack):
    return len(stack) == 0


def push(stack, x):
    stack.append(x)


def pop(stack):
    if isEmpty(stack):
        print("Error : stack underflow")
    else:
        return stack.pop()


'''prints element and NGE pair for all elements of
   arr[] '''


def printNGE(arr):
    s = createStack()
    element = 0
    next = 0

    # push the first element to stack
    push(s, arr[0])

    # iterate for rest of the elements
    for i in range(1, len(arr), 1):
        next = arr[i]

        if isEmpty(s) == False:

            # if stack is not empty, then pop an element from stack
            element = pop(s)

            '''If the popped element is smaller than next, then
                a) print the pair
                b) keep popping while elements are smaller and
                   stack is not empty '''
            while element < next:
                print(str(element) + " -- " + str(next))
                if isEmpty(s) == True:
                    break
                element = pop(s)

            '''If element is greater than next, then push
               the element back '''
            if element > next:
                push(s, element)

        '''push next to stack so that we can find
           next greater for it '''
        push(s, next)

    '''After iterating over the loop, the remaining
       elements in stack do not have the next greater
       element, so print -1 for them '''

    while isEmpty(s) == False:
        element = pop(s)
        next = -1
        print(str(element) + " -- " + str(next))


# Driver code
arr = [11, 13, 21, 3]
printNGE(arr)
````
